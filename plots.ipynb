```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [10, 20, 15, 25]
```

```python
plt.plot(x, y)
plt.show()
```

```python
x = ['Mon', 'Tues', 'Wed', 'Thur', 'Fri']
y = [10, 15, 7, 20, 12]

plt.plot(x, y, color='orange', linewidth=2, marker='o')
plt.title("Bakery Sales this Week")
plt.xlabel('Days of the Week')
plt.ylabel('Sales per Day')
plt.show()
```

```python
months = [1, 2, 3, 4, 5]
sales = [1000, 1500, 1200, 1600, 2000]
# plt.figure(figsize=(6,5))
plt.plot(months, sales, color='blue', linestyle=':', linewidth=2, 
         marker='o', label='2025 Sales Data')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Sales and Months Data')
plt.xticks(months, ['M1', 'M2', 'M3', 'M4', 'M5'])
plt.legend()
plt.grid()
plt.show()
```

```python
product = ['A', 'B', 'C', 'D']
sales = [1000, 1500, 800, 1200]

plt.bar(product, sales, color='orange', label='Sales 2025')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Product Sale Comparasion')
plt.legend()
plt.show()
```

```python
plt.barh(product, sales, color='orange', label='Sales 2025')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Product Sale Comparasion')
plt.legend()
plt.show()
```

```python
regions = ['North', 'South', 'East', 'West']
revenue = [3000, 2000, 1500, 1000]

plt.pie(revenue, labels=regions, autopct='%1.1f%%')
plt.legend()
plt.title("Revenue Contribution by region")
plt.show()
```

```python
scores = [45, 67, 89, 56, 78, 88, 92, 60, 74, 81, 59, 66, 75, 82, 90, 85, 70, 73, 68, 77]

plt.hist(scores, bins=5, color='Purple', edgecolor='black')
plt.xlabel('Score Range')
plt.ylabel("Number of Students")
plt.title("Score Distribution of Students")
plt.show()
```

```python
hours_studied = [1, 2, 3, 4, 5, 6, 7, 8]
exam_scores = [50, 55, 60, 65, 70, 75, 80, 85]

plt.scatter(hours_studied, exam_scores, color='orange', marker='^', label='Student Data')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Scores')
plt.title('Marks with studied hours')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.scatter([1, 2, 3], [50, 60, 70], color='blue', marker='o', label='Class A')
plt.scatter([1, 2, 3], [45, 50, 52], color='orange', marker='o', label='Class B')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Scores')
plt.title('Comparison of 2 Classes')
plt.legend()
plt.grid()
plt.show()
```
